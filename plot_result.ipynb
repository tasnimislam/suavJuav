```python
result = pd.read_csv("../model_result_iid_user_no_5db.csv")
plt.plot(list(result['User']), result['Original'], color = 'black', marker='o', linestyle='dashed', label = 'FedRec paper')
plt.plot(list(result['User']), result['Compensate'], color = 'red', marker='o', linestyle='dashed', label = 'Adding Compensation Layer')
plt.plot(list(result['User']), result['LSTM'], color = 'orange', marker='o', linestyle='dashed', label = 'RNN model')
plt.plot(list(result['User']), result['Average'], color = 'green', marker='o', linestyle='dashed', label = 'AveragePool')
plt.xlabel('Eb0/N0')
plt.ylabel('BER')
plt.title(f'plot ber for iid Eb0/N0 = 5 varying user number')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
plt.savefig(f'plot ber for iid Eb0/N0 = 5 varying user number.png')
```
